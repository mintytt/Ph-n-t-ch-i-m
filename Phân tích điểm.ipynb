```python
import pandas as pd
pd.set_option('display.max_rows', None)
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df1= pd.read_excel(r'C:\Users\VCB\Documents\Diem\Merge63.xlsx')
```

```python
df1.info()
```

```python
for column in df1.drop(['Tên trường lớp 12_x','Tên trường lớp 12_y'],axis=1).columns:
    print(f"===============Column: {column}==============")
    print(f"Number of unique values: {df1[column].nunique()}")
    print(f"Max: {df1[column].max()}")
    print(f"Min: {df1[column].min()}")
```

```python
# Visulazing the distibution of the data for every feature
df1.drop(['MSSV'],axis=1).hist(edgecolor='black', linewidth=1.2, figsize=(20, 20));
```

```python
plt.figure(figsize=(20, 20))
sns.heatmap(df1.drop(['MSSV'],axis=1).corr(), annot=True)
```

```python
df1.drop(['MSSV'],axis=1).describe()
```

```python
df1["Tên trường lớp 12_x"].value_counts().plot(kind='bar',figsize=(20,20))
```
